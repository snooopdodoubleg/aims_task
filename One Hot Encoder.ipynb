```python
import pandas as pd
import numpy as np
```

```python
dict1={"hello":["hola","hi","hi","henlo","bonjour","hi","henlo","henlo","bonjour","hola","hi"]}
df=pd.DataFrame(dict1)
df
```

```python
def MyOneHotEncoder(df,col):
    unique_vals=df[col].unique()
    # ham basiclly data frame ka basic structure bana rahe hai jse ham bad main df main dal de 
    dict1={i:[] for i in unique_vals} 
    for val in df[col]:
        if val in unique_vals:
             for i in dict1:
                if val==i:
                    dict1[i].append(1)
                else:
                    dict1[i].append(0)
    df.drop([col],inplace=True, axis=1)
    for i in dict1:
         df[i]=dict1[i]
    return df
MyOneHotEncoder(df,"hello")
```
